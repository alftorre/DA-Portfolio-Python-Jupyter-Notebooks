# Explore Python Syntax

## Function Example

```python
def to_celsius(x):
    '''Convert Farenheit to Celsius'''
    return (x-32) * 5/9

to_celsius(75)
```

## Conditional Example

```python
number = -4

if number > 0:
    print('Number is positive.')
elif number == 0:
    print('Number is 0')
else:
    print('Number is negative.')
    
```
